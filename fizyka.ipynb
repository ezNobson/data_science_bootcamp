```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import linregress
X=np.array([ 9.81,19.62,29.43,39.24,49.05,58.86,68.67,78.48,88.29 ,98.10 ])
Y=np.array([ 0.46,0.90,1.27,1.62,1.95,2.17,2.58,2.90,3.18,3.46 ])

slope, intercept, r_value, p_value, std_err = linregress(X, Y)

# Generate predicted values
predicted_Y = slope * X + intercept

# Plot the data and regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Dane pomiarowe')
plt.plot(X, predicted_Y, color='red', label=f'Regresja liniowa: y = {slope:.4f}x + {intercept:.4f}')

plt.title("Wykres: Wyduenie rednie od Siy")
plt.xlabel("Sia [N]")
plt.ylabel("Wyduenie rednie [mm]")
plt.grid(True)
plt.legend()
plt.show()
print(f"Niepewno wspczynnika kierunkowego (u(a)): {std_err:.4f}")
```
